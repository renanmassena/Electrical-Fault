```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
nb_version = "v2"
test_split = 0.3
scaler = MinMaxScaler()
```

```python
df = pd.read_csv('../../data/raw/detection_data.csv')
df = df.drop(['Unnamed: 7','Unnamed: 8'], axis=1)
df = df.rename(columns={'Output (S)':'output'})
df
```

```python
X = df.drop('output', axis=1)
Y = df['output']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_split, random_state=42, stratify=Y)
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
df_train = pd.concat([Y_train.reset_index(), X_train_scaled], axis=1)
df_test = pd.concat([Y_test.reset_index(), X_test_scaled], axis=1)
display(f'Train: {df_train.shape}', df_train.head(), f'Test: {df_test.shape}', df_test.head())
```
